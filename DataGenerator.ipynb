```python
from pathlib import Path
import os
import shutil
```

```python
fdata = f'../../data/sars-cov2-ct-scan'
val_split = 0.1
test_split = 0.2
train_split = 1 - test_split
```

```python
p = Path(fdata)
ptrain = p / 'train'
ptest = p / 'test'
pval = p / 'val'
for item in ptrain.iterdir():
    Path(ptest / item.stem).mkdir(parents=True, exist_ok=True)
    Path(pval / item.stem).mkdir(parents=True, exist_ok=True)
```

```python
# Train to Test
for item in ptrain.iterdir():
    ntrain = len(os.listdir(item))
    ntot = int(ntrain * test_split)
    print(f'len(ntrain): {ntrain}, ntot: {ntot}')
    for sample in item.iterdir():
        print(f'{sample} to {ptest / item.stem / sample.name}')
        shutil.move(sample, ptest / item.stem / sample.name)
        ntot = ntot-1
        if ntot == 0:
            break
    print(f'ntot: {ntot}')
```

```python
# Train to Val
for item in ptrain.iterdir():
    ntrain = len(os.listdir(item))
    ntot = int(ntrain * val_split)
    print(f'len(ntrain): {ntrain}, ntot: {ntot}')
    for sample in item.iterdir():
        print(f'{sample} to {pval / item.stem / sample.name}')
        shutil.move(sample, pval / item.stem / sample.name)
        ntot = ntot-1
        if ntot == 0:
            break
    print(f'ntot: {ntot}')
```
